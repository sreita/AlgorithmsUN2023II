# **BubbleSortOpt Distribution Without Repetitions**

```python
import matplotlib.pyplot as plt
from itertools import permutations
import numpy as np
import time

def bubble_sort_opt_counters(arr):
    comparisons = 0
    swaps = 0
    passes = 0

    n = len(arr)
    for i in range(n - 1):
        swapped = False
        for j in range(0, n - i - 1):
            comparisons += 1
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
                swapped = True
        passes += 1
        if not swapped:
            break

    return comparisons, swaps, passes

def generate_permutations_without_repetitions(n):
    elements = list(range(1, n + 1))
    perms = permutations(elements)
    return list(perms)

def plot_time_histogram(n):
    permutations = generate_permutations_without_repetitions(n)
    time_list = []

    for perm in permutations:
        start_time = time.time()
        bubble_sort_opt_counters(list(perm))
        end_time = time.time()
        elapsed_time = end_time - start_time
        time_list.append(elapsed_time)

    #print("Time Results: ", time_list)

    # Convertir tiempos a notación científica con una sola décima
    time_list_str = [f"{t:.1e}" for t in time_list]

    # Obtener valores únicos y sus frecuencias
    unique_times, frequencies = np.unique(time_list_str, return_counts=True)

    plt.figure(figsize=(10, 6))
    plt.bar(unique_times, frequencies, color='orange', edgecolor='black')
    plt.title('Distribution of Time')
    plt.xlabel('Time (e-06 seconds)')
    plt.ylabel('Frequency')

    # Configuración de los ticks del eje x para mostrar valores ordenados ascendente
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    plt.show()

    print(unique_times)

def plot_data_bubbleSort(n):
    permutations = generate_permutations_without_repetitions(n)
    num_permutations = len(permutations)

    comps_list = []
    swaps_list = []
    steps_list = []
    passes_list = []

    for perm in permutations:
        comps, swaps, passes = bubble_sort_opt_counters(list(perm))
        comps_list.append(comps)
        swaps_list.append(swaps)
        steps_list.append(comps + swaps)
        passes_list.append(passes)

    b = np.array(range(min(comps_list)-1,max(comps_list)+2) ) +0.5
    plt.hist(comps_list,bins=b, color="red")
    plt.title("Bubble Sort Comparisons")
    plt.xlabel("Comparisons ")
    plt.ylabel("Frequency")
    plt.show()

    b = np.array(range(min(steps_list)-1,max(steps_list)+2) ) +0.5
    plt.hist(steps_list,bins=b,color="green")
    plt.title("Bubble Sort Steps")
    plt.xlabel("Steps ")
    plt.ylabel("Frequency")
    plt.show()

    b = np.array(range(min(passes_list)-1,max(passes_list)+2) ) +0.5
    plt.hist(passes_list,bins=b,color="brown")
    plt.title("Bubble Sort Passes")
    plt.xlabel("Passes")
    plt.ylabel("Frequency")
    plt.show()
```

```python
n = 3


print("n=3")
plot_time_histogram(n)
plot_data_bubbleSort(n)
```

# **BubbleSortOpt Distribution With Repetitions**

```python
from itertools import product

def plot_time_histogram_with_repetitions(n):
    permutations = product(range(1,n+1), repeat = n)
    time_list = []

    for perm in permutations:
        start_time = time.time()
        bubble_sort_opt_counters(list(perm))
        end_time = time.time()
        elapsed_time = end_time - start_time
        time_list.append(elapsed_time)

    #print("Time Results: ", time_list)

    # Convertir tiempos a notación científica con una sola décima
    time_list_str = [f"{t:.1e}" for t in time_list]

    # Obtener valores únicos y sus frecuencias
    unique_times, frequencies = np.unique(time_list_str, return_counts=True)

    plt.figure(figsize=(10, 6))
    plt.bar(unique_times, frequencies, color='orange', edgecolor='black')
    plt.title('Distribution of Time')
    plt.xlabel('Time (e-06 seconds)')
    plt.ylabel('Frequency')

    # Configuración de los ticks del eje x para mostrar valores ordenados ascendente
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

    plt.show()

    print(unique_times)

def plot_data_bubbleSort_with_repetitions(n):
    permutations = list(product(range(1,n+1), repeat = n))
    num_permutations = len(permutations)

    comps_list = []
    swaps_list = []
    steps_list = []
    passes_list = []

    for perm in permutations:
        comps, swaps, passes = bubble_sort_opt_counters(list(perm))
        comps_list.append(comps)
        swaps_list.append(swaps)
        steps_list.append(comps + swaps)
        passes_list.append(passes)

    b = np.array(range(min(comps_list)-1,max(comps_list)+2) ) +0.5
    plt.hist(comps_list,bins=b, color="red")
    plt.title("Bubble Sort Comparisons")
    plt.xlabel("Comparisons ")
    plt.ylabel("Frequency")
    plt.show()

    b = np.array(range(min(steps_list)-1,max(steps_list)+2) ) +0.5
    plt.hist(steps_list,bins=b,color="green")
    plt.title("Bubble Sort Steps")
    plt.xlabel("Steps ")
    plt.ylabel("Frequency")
    plt.show()

    b = np.array(range(min(passes_list)-1,max(passes_list)+2) ) +0.5
    plt.hist(passes_list,bins=b,color="brown")
    plt.title("Bubble Sort Passes")
    plt.xlabel("Passes")
    plt.ylabel("Frequency")
    plt.show()
```

# **Fibonacci**

```python
def fibRecursive(n):
    if n==0 :
        return 0
    if n==1 :
        return 1
    return fibRecursive(n-1)+fibRecursive(n-2)

def fibIterative(n):
    a, b = 0, 1
    for _ in range(n):
        a, b = b, a + b
    return a

def fibIterativeLogaritmic(n):
   if n==0: return 0


   v1, v2, v3 = 1, 1, 0

   for rec in bin(n)[3:]:
        calc = (v2*v2)
        v1, v2, v3 = (v1*v1+calc) , ((v1+v3)*v2) , (calc+v3*v3)
        if rec == '1': v1, v2, v3 = (v1+v2), v1, v2
   return v2
```

```python

fibRecursiveTime=[]
fibIterativeTime=[]

fibIterativeLogaritmicTime=[]


for i in range (0,43):
    begin = time.time()
    fibRecursive(i)
    end = time.time()
    timeElapsed = end-begin
    fibRecursiveTime.append(timeElapsed)

for i in range (0,101):
    begin = time.time()
    fibIterative(i)
    end = time.time()
    timeElapsed = end -begin
    fibIterativeTime.append(timeElapsed)


for i in range (0,101):
    begin = time.time()
    fibIterative(i)
    end = time.time()
    timeElapsed = end - begin
    fibIterativeLogaritmicTime.append(timeElapsed)



b = np.array(range(0,43) )
c = np.array(range(0,101))


plt.plot(b,fibRecursiveTime)
plt.title("Recursive time given n")
plt.xlabel("n size")
plt.ylabel("Seconds")
plt.show()

plt.plot(c,fibIterativeTime)
plt.title("Iterative time given n")
plt.xlabel("n size")
plt.ylabel("Seconds")
plt.show()

plt.plot(c,fibIterativeLogaritmicTime)
plt.title("Iterative logaritmic time given n")
plt.xlabel("n size")
plt.ylabel("Seconds")
plt.show()

```
